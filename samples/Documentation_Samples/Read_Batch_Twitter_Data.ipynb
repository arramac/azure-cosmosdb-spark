```
%sql
select 
  id,
  created_at,
  screen_name,
  location,
  text,
  retweet_count,
  hashtags,
  user_mentions,
  favorited,
  source
from tweets
limit 20
```

```
%sql
select hashtags.text, count(distinct id) as tweets
from (
select 
  explode(hashtags) as hashtags,
  id
from tweets
) a
group by hashtags.text
order by tweets desc
limit 10
```
